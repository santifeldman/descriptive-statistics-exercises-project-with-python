````python
import numpy as np
import statistics as stats
from scipy.stats import kurtosis, skew

def calculate_statistics(array):
    statistics = {
        'media': stats.mean(array),
        'mediana': stats.median(array),
        'moda': stats.mode(array),
        'varianza': stats.variance(array),
        'desvio estandar': stats.stdev(array),
        'asimetria': skew(array),
        'curtosis': kurtosis(array)
    }
    return statistics

def print_statistics(statistics, array_name):
    print(f'Estad√≠sticas para {array_name}:')
    print(f"La media del array es de {statistics['media']}")
    print(f"La mediana es de {statistics['mediana']}")
    print(f"La moda es de {statistics['moda']}")
    print(f"La varianza del array es de {statistics['varianza']}")
    print(f"El desvio estandar es {statistics['desvio estandar']}")
    print(f"La asimetria es de {statistics['asimetria']}")
    print(f"La curtosis es de {statistics['curtosis']}")
    print()

# Generar array1 con distribuci√≥n normal
array1 = np.random.normal(loc=0.0, scale=1.0, size=100)
stats_array1 = calculate_statistics(array1)
print(array1)
print_statistics(stats_array1, "array1")

# Generar array2 con distribuci√≥n chi-cuadrado
array2 = np.random.chisquare(df=3, size=100)
stats_array2 = calculate_statistics(array2)
print(array2)
print_statistics(stats_array2, "array2")
````

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import statistics as stats  
data = [4,2,5,8,6]
std = stats.stdev(data)
print(f"Desviaci√≥n est√°ndar: {std}")


import math
def calcular_media(data):
    return sum(data) / len(data)

def calcular_desvio_estandar(data):
    media = calcular_media(data)
    varianza = sum((x - media) ** 2 for x in data) / (len(data) -1)
    desvio_estandar = math.sqrt(varianza)
    return desvio_estandar

desvio_estandar = calcular_desvio_estandar(data)
print (desvio_estandar)
````
